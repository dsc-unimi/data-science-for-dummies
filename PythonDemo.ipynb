<b> MFR in action </b>

```python
#Prodotto fra il triplo dei numeri da 1 a 5 che risulta pari
l = [i for i in range(1,6)]
ltriple = list(map(lambda x: 3*x,l))
evens = list(filter(lambda x: x%2==0,ltriple))

from functools import reduce
product = reduce((lambda x, y: x * y), evens)
product
```

Per capire meglio il passaggio di funzioni come parametri, osserviamo una semplice versione del pattern visitor:

```python
def visit(data,visitor=print):
    for x in data:
        visitor(x)

visit([1,2,3,4,5],print)
```

- Quanti sono i pokémon di prima generazione che hanno la somma delle statistiche di base maggiori della media?

```python
strong_pokemon=len(first_gen[first_gen['base_total']>=first_gen['base_total'].mean()])
weak_pokemon=len(pokemon)-strong_pokemon
print("Strong pokemon: " + str(strong_pokemon) + "\n" + "Weak_pokemon: " + str(weak_pokemon) +"\n")
```

Visualizziamo l'informazione attraverso un grafico:

```python
strong_weak=[strong_pokemon,weak_pokemon]
labels=['Strong Pokemon',"Weak Pokemon"]
sw=pd.DataFrame({'': strong_weak},index=labels)
sw.plot.pie(y='', figsize=(5, 5))
plt.show()
```

- Ci sono più pokemon maschi o più pokémon femmina?

```python
def max_gender(pmale,pfemale):
    gender=""
    if not(pmale) and not(pfemale):
        gender="None"
    elif not(pfemale):
        gender="Male"
    elif not(pmale):
        gender="Female"
    elif pfemale>=pmale:
        gender="Female"
    else:
        gender="Male"
    return gender
```

```python
percentage_males = [p if p else None for p in pokemon['percentage_male']]
percentage_females = [(100-p) if p else None for p in pokemon['percentage_male']]

prevalent_genders = []
for i in range(0,len(percentage_females)):
    prevalent_genders.append(max_gender(percentage_males[i],percentage_females[i]))

pd.Series(prevalent_genders).value_counts()
```
